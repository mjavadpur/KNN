```python
dictionary = {'age' : [] , 'sex' : [] , 'cp' : [] , 'trtbps' : [] , 'chol' : [] , 'fbs' : [] , 'restecg' : [] , 'thalachh' : [] ,
       'exng' : [] , 'oldpeak' : [] , 'slp' : [] , 'caa' : [] , 'thall' : [] }
df_scaled = pd.DataFrame(dictionary)

for i in ['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh',
       'exng', 'oldpeak', 'slp', 'caa', 'thall']:
       array = []
       mean = heart_df[i].mean()
       std = heart_df[i].std()
       for x in heart_df[i]:
        array.append((x-mean)/std)
       df_scaled[i] = array

df_scaled
```

New data STANDARDIZATION

```python
testDataScaled = pd.DataFrame({'age':[], 'sex':[], 'cp':[], 'trtbps':[], 'chol':[], 'fbs':[], 'restecg':[], 'thalachh':[],
       'exng':[], 'oldpeak':[], 'slp':[], 'caa':[], 'thall':[]})

for i in testData.columns:
       mean = df[i].mean()
       std = df[i].std()
       testDataScaled[i] = ((d[i]-mean)/std)


testDataScaled
```

```python
import numpy as np
```

```python
np_scaled_test = testDataScaled.to_numpy()
```

```python
np_df_scaled = df_scaled.to_numpy()
```

```python
euclidean_distance = np.zeros(len(np_df_scaled))

for i in range(len(np_df_scaled)):
  # euclidean_distance[i] = np.sqrt(np.sum(np.power(np_df_scaled[i]-np_scaled_test,2)))
  euclidean_distance[i] = np.sqrt(np.sum([(a-b)*(a-b) for a , b in zip(np_df_scaled[i], np_scaled_test)]))
```

```python
df_scaled['euclidean_distance'] = euclidean_distance
```

```python
df_scaled['output'] = df['output']
df_scaled
```

```python
sorted_df = df_scaled.sort_values(by=['euclidean_distance'])
sorted_df
```
