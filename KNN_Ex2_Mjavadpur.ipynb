یک دیتا فریم با پنج ستون و هر ستون شامل ۱۰۰ عدد رندم در بازه های مختلف (مثلا یکی ۱۰۰۰ تا ۵۰۰۰ یکی دیگه ۵۰۰- تا ۱۰۰ و ...) بسازید و داده را استانداردسازی کنید در آخر نشان دهید که میانگین صفر و انحراف معیار داده استاندارد سازی شده یک است.


```python
import pandas as pd
import random as rnd


dictionary = {'Col1' : [] , 'Col2' : [] , 'Col3' : [] , 'Col4' : [] , 'Col5' : [] }
df = pd.DataFrame(dictionary)


for i in dictionary:
  z = 1
  num = rnd.randint(100*z,300*z)
  array = []
  for j in range(0,100):
    array.append(num*rnd.random())
  df[i] = array
  z *=10

print("Scaled DataFrame shape is: \n" + str(df.shape))
print("\nDataFrame mean is: \n"+ str(df.mean()))
print("\nDataFrame STD is: \n"+ str(df.std()))

df_scaled = pd.DataFrame(dictionary)
for i in dictionary:
       array = []
       mean = df[i].mean()
       std = df[i].std()
       for x in df[i]:
        array.append((x-mean)/std)
       df_scaled[i] = array


print("\n\nScaled DataFrame shape is: \n" + str(df_scaled.shape))
print("\nScaled DataFrame mean is: \n"+ str(df_scaled.mean().round(2)))
print("\nScaled DataFrame STD is: \n"+ str(df_scaled.std()))


```
